```python
import matplotlib.pyplot as plt
import math
import numpy as np
import sympy as sp

def f(x):
    return x**2+4*math.sin(x)

x_sym = sp.Symbol('x')
f_sym = x_sym**2+4*sp.sin(x_sym)

a = -3
b = 3

c = (a+b)/2
```

```python
diff1 = f_sym.diff()
diff2 = diff1.diff()
diff3 = diff2.diff()
diff4 = diff3.diff()
diff5 = diff4.diff()
```

```python
def Q1(x, c):
    return f_sym.subs(x_sym, c).evalf() + diff1.subs(x_sym, c).evalf()*(x-c)
    
def Q2(x, c):
    return Q1(x, c) + diff2.subs(x_sym, c).evalf()*((x-c)**2)/(math.factorial(2))

def Q3(x, c):
    return Q2(x, c) + diff3.subs(x_sym, c).evalf()*((x-c)**3)/(math.factorial(3))

def Q4(x, c):
    return Q3(x, c) + diff4.subs(x_sym, c).evalf()*((x-c)**4)/(math.factorial(4))

def Q5(x, c):
    return Q4(x, c) + diff5.subs(x_sym, c).evalf()*((x-c)**5)/(math.factorial(5))
```

```python
x = np.arange(a,b,0.001)
y = [f(i) for i in x]
```

```python
Q1_ = [Q1(i, c) for i in x]
plt.plot(x, y, label='f')
plt.plot(x, Q1_, label='Q1')
plt.legend()
plt.show()
```

```python
def geo_mean(iterable):
    a = np.array(iterable)
    return a.prod()**(1.0/len(a))
```

```python
xs = [-2.5, -1, 0, 2, 3]
ys = [f(i) for i in xs]
```

```python
def Lagr(x, i, xs, ys):
    N = len(xs)-1
    L = 1
    for j in range(0, N):
        if j != i:
            L = L*(x-xs[j])/(xs[i]-xs[j]) 
    L = L*ys[i]
    return L

def Lagrang(x, N, xs, ys):
    s = 0
    for i in range(N):
        s += Lagr(x, i, xs, ys)
    return s
```

```python
x = np.arange(a,b,0.001)
y = [f(i) for i in x]
```

```python
y_l1 = [Lagrang(i, 1, xs, ys) for i in x]
plt.plot(x,y, label='f')
plt.plot(x, y_l1, label='Lagr')
plt.legend()
print(max(abs((np.array(y) - np.array(y_l1)))))
```

```python
y_l2 = [Lagrang(i, 2, xs, ys) for i in x]
plt.plot(x,y, label='f')
plt.plot(x, y_l2, label='Lagr')
plt.legend()
print(max(abs((np.array(y) - np.array(y_l2)))))
```
